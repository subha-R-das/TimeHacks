```python
import pandas as pd

# Assuming df_3405 is your DataFrame and it contains 'temperature', 'humidity', 'pm10', 'date', and 'device_id' columns

# Function to apply all the changes for a single device_id
def process_device_data(df):
    df.sort_values(by='timestamp', inplace=True)
    df['temp_change'] = df['temperature'].diff().ne(0).cumsum()
    df['humid_change'] = df['humidity'].diff().ne(0).cumsum()
    df['pm10_change'] = df['pm10'].diff().ne(0).cumsum()
    df['pm2_5_change'] = df['pm2_5'].diff().ne(0).cumsum()
    def replace_constant_temps(group):
      counts = group.groupby('temp_change').size()
      to_replace = counts[counts > 5].index.tolist()
      group.loc[group['temp_change'].isin(to_replace), 'temperature'] = 0
      return group
    def replace_constant_humid(group):
      counts = group.groupby('humid_change').size()
      to_replace = counts[counts > 5].index.tolist()
      group.loc[group['humid_change'].isin(to_replace), 'humidity'] = 0
      return group
    def replace_constant_pm10(group):
      counts = group.groupby('pm10_change').size()
      to_replace = counts[counts > 5].index.tolist()
      group.loc[group['pm10_change'].isin(to_replace), 'pm10'] = 0
      return group
    def replace_constant_pm2_5(group):
      counts = group.groupby('pm2_5_change').size()
      to_replace = counts[counts > 5].index.tolist()
      group.loc[group['pm2_5_change'].isin(to_replace), 'pm2_5'] = 0
      return group

    df = df.groupby('date').apply(replace_constant_temps)
    df = df.groupby('date').apply(replace_constant_humid)
    df = df.groupby('date').apply(replace_constant_pm10)
    df = df.groupby('date').apply(replace_constant_pm2_5)

    # Drop the change tracking columns
    df.drop(['temp_change', 'humid_change', 'pm10_change','pm2_5_change'], axis=1, inplace=True)
    return df

# Separate the DataFrame by device_id and process each separately
list_processed_dfs = []
for device_id, group in df_sorted.groupby('device_id'):
    processed_df = process_device_data(group.copy())
    list_processed_dfs.append(processed_df)

# Concatenate all the processed DataFrames back together
final_df = pd.concat(list_processed_dfs).reset_index(drop=True)

# final_df now contains the processed data for all devices
```
